**Calculo Volatilidad**

```python
def rendimiento(df_final_Cafe):
    df_final_Cafe["Rendimiento"] =  (df_final_Cafe["precio_cafe"].shift(-1) - df_final_Cafe["precio_cafe"])/ df_final_Cafe["precio_cafe"]
    return df_final_Cafe
```

```python
df_final_Cafe = rendimiento(df_final_Cafe)
df_final_Cafe
```

```python
df_final_Cafe['Promedio rendimiento'] = df_final_Cafe.groupby('año')['Rendimiento'].transform('mean')
df_final_Cafe
```

```python
def calcular_diferencia_r(df_final_Cafe):
    df_final_Cafe["Diferencia cuadrado"] = (df_final_Cafe["Rendimiento"]-df_final_Cafe['Promedio rendimiento'])**2
    return df_final_Cafe

# Calculamos la nueva columna usando transform
df_final_Cafe = calcular_diferencia_r(df_final_Cafe)
df_final_Cafe
```

```python
df_final_Cafe['Suma diferencia cuadrados'] = df_final_Cafe.groupby('año')['Diferencia cuadrado'].transform('sum')
df_final_Cafe
```

```python
def volatilidad(df_final_Cafe):
    df_final_Cafe["Volatilidad"] = np.sqrt((df_final_Cafe['Suma diferencia cuadrados'])/11)
    return df_final_Cafe

# Calculamos la nueva columna usando transform
df_final_Cafe = volatilidad(df_final_Cafe)
df_final_Cafe
```

```python
df_Cafe = df_final_Cafe[["año","fecha","indice_dolar","precio_petroleo","costo_produccion_cafe","precio_cafe","Volatilidad"]]
df_Cafe
```
